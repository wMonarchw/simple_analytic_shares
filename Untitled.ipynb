```python
gspc = yf.download('^GSPC', '2022-01-01')
```

```python
daily_returns = stock['Adj Close'].pct_change().dropna()

gspc_returns = gspc['Adj Close'].pct_change().dropna()

joined_data = pd.concat([daily_returns, gspc_returns], axis=1, join='inner')

cov = joined_data.cov()

cov_vs_market = cov.iloc[:-1, -1]

sp_var = gspc_returns.var() * 252

beta = cov_vs_market / sp_var
print(beta)
```

Теперь можно расчитать VAR или же Value At Risk

```python
def var(Position, sigma, Quantile):
    return Position * sigma * scipy.stats.norm.ppf(Quantile)
```

```python
daily_returns = stock['Adj Close'].pct_change()
```

```python
daily_returns = daily_returns.dropna()
```

```python
daily_returns
```

```python
# Стандартное отклонение
sigma = np.std(daily_returns, axis=0)
```

```python
sigma
```

Предположим что наша Position = 1000$ в акциях

```python
Position = 1000  
Quantile = 0.95


variances = {}
for ticker in sigma.index:
    variances[ticker] = var(Position, sigma[ticker], Quantile)

print("VAR:")
print(variances)
```

Теперь предположим что мы просто покупаем последнюю позицию со всеми акциями и сделаем расчеты основываясь на этом.

```python
position = stock['Adj Close'].iloc[-1]
```

```python
Vararray = []
for ticker in position.index:
    Vararray.append(var(position[ticker], sigma[ticker], 0.95))
```

Теперь хотелось бы рассмотреть Stock Market Predictions (Предсказание стоимости) используя линейную регрессию.

```python
import statsmodels.api as sm
from statsmodels.tsa.ar_model import AutoReg, ar_select_order
```

```python
stock_time = stock.asfreq('d')
stock_time = stock_time.bfill()
```

```python
stock_time = stock_time['Adj Close']


print(stock_time.head())


stock_time.plot(figsize=(15, 6))
plt.title('Adjusted Close Prices')
plt.xlabel('Date')
plt.ylabel('Price')
plt.show()
```

```python
stock_time
```

```python
from statsmodels.tsa.arima.model import ARIMA
```

```python
def AutoReg_forecast(df, ticker):
    series = df[ticker]

    model = AutoReg(series, lags=1)  
    model_fit = model.fit()

    forecast = model_fit.predict(start=len(series), end=len(series)+29)  

    return forecast
```

```python
stock_data = stock_time.unstack()
```

```python
forecast_data = {}
ticker_list = ['TSLA', 'AAPL', 'MSFT', 'BABA', 'NVO']
for ticker in ticker_list:
    forecast_data[ticker] = AutoReg_forecast(stock_data, ticker)
```

```python
plt.figure(figsize=(10, 6))
```

```python
for ticker in ticker_list:
    plt.plot(stock_data[ticker], label=ticker) 
    plt.plot(forecast_data[ticker], label=ticker + ' (Forecast)', color='black')  # Сделаем черным Forecast 

plt.title('Prices with Forecast')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.show()
```
